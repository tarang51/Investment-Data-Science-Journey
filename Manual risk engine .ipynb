```python
Daily_returns=[1.22,1.29,2.24,3.32,4.25,8.25,6.25,5.25,4.36,9.25,7.82,6.29,4.12,8.35,1.95,6.23,1.42,10.00,9.95,10.25]
print(Daily_returns)
```

```python
# crete a funciton to calculate volatility (Daily_returns)
def calculate_volatility (Daily_returns):
    mean=sum (Daily_returns)/len(Daily_returns)
    squared_diff= [(x-mean)**2 for x in Daily_returns]
    variance= sum(squared_diff)/len(squared_diff)
    volatility= (variance)**0.5
    return (volatility)

print(f"The volatility is: {calculate_volatility(Daily_returns)}")
 
```

```python
first_half=Daily_returns[:10]
second_half=Daily_returns[10:]
print(first_half,second_half)
```

```python
# 2. Use your existing function to get volatility for both
vol_first = calculate_volatility(first_half)
vol_second = calculate_volatility(second_half)

# 3. Print and Compare
print(f"First 10 days Volatility: {vol_first:.4f}")
print(f"Last 10 days Volatility: {vol_second:.4f}")
```
